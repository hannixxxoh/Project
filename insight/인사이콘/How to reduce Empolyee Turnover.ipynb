## 경력이 짧은 신입

### EnvironmentSatisfaction

```python
#크로스태블릿
pd.crosstab(data.EnvironmentSatisfaction,
            data.Attrition, margins=True).style.background_gradient(
                cmap='summer_r')  # margins=True => All 출력
```

```python
#팩터플랏

sns.factorplot('EnvironmentSatisfaction', 'Attrition', data=data)
plt.show()
```

```python
sns.barplot('EnvironmentSatisfaction', 'Attrition', data=data)
plt.title('EnvironmentSatisfaction vs Attrition')
```

* 만족도 2 이상의 직원들은 모두 0.15 이하로 유사한 퇴사율을 보였지만, 만족도 1인 직원들은 0.25가 넘는 비교적 높은 퇴사율을 보였음
* 이는 Very HIgh(4점)과 High/Medium(2~3점)의 경우 퇴사까지 결정하지 않고 회사를 계속 다니지만, Low(1점)의 경우 퇴사를 결정한다는 의미.
* 직원들의 환경만족도가 __완전 불만족까지 떨어지지 않도록 유지__하는 것이 관건임을 알 수 있음.

### WorkLifeBalance

```python
#팩터플랏
sns.factorplot('WorkLifeBalance', 'Attrition', data=data) ## 먼저 그래프를 그린 후
fig = plt.gcf() ## 현재 figure에 대해서
fig.set_size_inches(5, 5) ## figure 사이즈 설정
plt.show()
```

```python
sns.barplot('WorkLifeBalance', 'Attrition', data=data, ci=None)
plt.title('WorkLifeBalance vs Attrition')
```

* 이때, Good(1점), Better(3점), Best(4점)의 경우 0.15 내외의 유사한 퇴사율을 보이나, Bad(1점)는 두 배 남짓의 높은 퇴사율을 보임.
* 환경만족도와 마찬가지로 직원들의 __워라밸만족도가 완전 불만족까지 떨어지지 않도록 유지__하는 것이 관건임을 알 수 있음.

### MonthlyIncome&MonthlyRate

```python
data['MonthlyRate_band'] = pd.qcut(data['MonthlyRate'], 4, labels=[0, 1, 2, 3]) #네개로 나눠주겠다
sns.barplot('MonthlyRate_band', 'Attrition', data=data, ci=None)
plt.title('MonthlyRate_band vs Attrition')
```

* MonthlyRate에 따른 퇴사율은 0.144에서 0.179로 큰 차이가 없음.
* MonthlyRate이 직원들의 퇴사 결정에 큰 영향을 주지 못함을 의미한다.

```python
data['MonthlyIncome_band'] = pd.qcut(data['MonthlyIncome'], 4, labels=[0, 1, 2, 3]) #네개로 나눠주겠다
sns.factorplot('MonthlyIncome_band', 'Attrition', data=data, ci=None)
plt.show()
```

* 반면, MonthlyIncome이 낮은 경우(0) 퇴사율은 0.293 수준으로 매우 높음.
* MonthlyIncome은 직원들의 퇴사 결정에 아주 중요한 변수임을 알 수 있다.
* 그렇다면 MonthlyIncome과 MonthlyRate 중 무엇이 중요할까?

```python
data.groupby(['MonthlyIncome_band','MonthlyRate_band'])['Attrition'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

* MonthlyIncome이 낮으면(0) 퇴사율이 0.252~0.324에 해당할 정도로 높다.
* 반면 MonthlyRate이 아무리 높더라도 MonthlyIncome이 낮으면 퇴사율이 높다
* __즉, 회사 차원의 추가적인 cost(복지, 보험 등)보다는 높은 임금이 동기부여가 됨!__
* __다시 말하면 기타 비용을 줄이고 직원 임금을 상향해야 함.__

### MonthlyIncome&PercentSalaryHike

```python
data['PercentSalaryHike_band'] = pd.qcut(data['PercentSalaryHike'], 3, labels=[0, 1, 2]) #네개로 나눠주겠다
data.groupby(['PercentSalaryHike_band'])['Attrition'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

* 연봉인상률이 낮으면 퇴사율이 높다.
* 그렇다면 MonthlyIncome과 PercentSalaryHike 중 무엇이 중요할까?

```python
data.groupby(['MonthlyIncome_band', 'PercentSalaryHike_band'])['Attrition'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

* Of course! MonthlyIncome!
* 연봉인상률이 아무리 높아도 월급이 낮으면 퇴사율을 낮출 수 없다. 반면 월급이 높으면 퇴사율도 낮아진다.
* 우리의 주요 퇴사자: 경력이 짧은 신입
* __즉, 초봉을 높이고(동기부여) 연봉인상률은 낮추는(cost 감소) 전략이 필요함.__

```python
sns.factorplot('PercentSalaryHike_band', 'Attrition', data=data, ci=None, hue='MonthlyIncome_band')
plt.show()
```

* MonthlyIncome이 낮으면 PercentSalaryHike가 높더라도 퇴사율이 높음. (red line)
* 반면 PercentSalaryHike가 낮아도 MonthlyIncome이 높으면 퇴사율이 낮아짐. (gray line)
* 이 전제는 WorkLifeBalance에도 적용되는 것을 알 수 있다.

### MonthlyIncome&WorkLifeBalance

```python
data.groupby(['MonthlyIncome_band', 'WorkLifeBalance'])['Attrition'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

* MonthlyIncome이 낮으면 WorkLifeBalance에 아무리 만족하더라도 퇴사율이 높다.
* 이때, MonthlyIncome도 낮은데다(0) WorkLifeBalance에 불만족하면(1) 퇴사율이 0.524로 굉장히 높다.

### MonthlyIncome&JobRole

```python
data.groupby(['JobRole'])['MonthlyIncome'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

* 퇴사율이 가장 높은 부서인 Sales Representative는 평균 MonthlyIncome이 제일 낮다.

## 개선 방안 <br>
**1. 초봉up 기타비용down**: 월급은 복지, 인센티브, 워라밸보다 퇴사를 결정하는 중요한 지표!<br>즉, 신입의 초봉을 높이고 기타비용을 낮추는 전략이 필요함.<br>
복지, 인센티브, 워라밸에 소요되는 비용을 줄이고 신입의 초봉에 집중!<br><br>__2. 개인과업에 따른 명확한 보상체계 구축__: 현재 출장과 초과근무를 수행하는 직원들의 MonthlyIncome이 낮음. <br>일한 만큼 정당한 보상을 받지 않고 있고, 이는 높은 퇴사율로 이어지고 있음. <br>즉, 회사는 사원들의 추가적인 노동에 상응하는 금전적 보상을 명확히 지급해야 함.
